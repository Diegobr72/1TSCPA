Crie gráficos das seguintes funções:



1.   \begin{align}\mathbf{x}  &= 2.5x + 3x^2 -0.8^x\end{align}

2.   \begin{align}\mathbf{x}  &= \sqrt{2x} + 1.7x\end{align}
3.   \begin{align}\mathbf{x}  &= \frac{x}{2}-8x\end{align}






```python
import matplotlib.pyplot as plt
import numpy as np

# Define a função f(x) = 2.5x + 3x² - 0.8^x
def funcao(x):
    return 2.5*x + 3*x**2 - 0.8**x

# Cria um array de valores de x para o gráfico
x_vals = np.linspace(-5, 5, 400)
y_vals = funcao(x_vals)

# Gera o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='$f(x) = 2.5x + 3x^2 - 0.8^x$')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')

# Adiciona rótulos e título
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de $f(x) = 2.5x + 3x^2 - 0.8^x$')

# Exibe o gráfico
plt.legend()
plt.grid(True)
plt.show()
```

# Limites

```python
from sympy import symbols, limit

# Define a variável simbólica
x = symbols('x')

# Define a expressão (x²-1)/(x-1)
f_x = (x**2 - 1)/(x-1)

#Calcula o limite quando x tende à 1
limite_tendendo_a_1 = limit(f_x, x, 1)
print("O limite da função quando se aproxima de 1 é: ", limite_tendendo_a_1 )
```

```python
x = symbols('x')
f_x = abs(x)/x

#Calcula o limite quando x tende à 0 pela esquerda
limite_esquerda = limit(f_x, x, 0, dir='-')

#Calcula o limite quando x tende à 0 pela direita
limite_direita = limit(f_x, x, 0, dir='+')

print("O limite da função |x|/x a medida que x se aproxima de zero pela esquerda é: ", limite_esquerda)
print("O limite da função |x|/x a medida que x se aproxima de zero pela direita é: ", limite_direita)
```

Vamos considerar um cenário onde queremos maximizar a receita de venda de um produto, dada uma função que modela a relação entre o preço de venda e a quantidade vendida.

Suponha que a receita (R) seja dada por R(x) = -2x² +100x, onde x é quantidade do produto vendida e R é a receita correspondente. Vamos encontrar a quantidade que maximiza a receita.

```python
from sympy import symbols, diff, solve

# Define a variável simbólica
x = symbols('x')

# Define a função de receita R(x) = -2x^2 + 100x
receita = -2*x**2 + 100*x

# Calcula a derivada de R em relação a x
derivada_receita = diff(receita, x)

# Resolve a equação derivada = 0 para encontrar pontos críticos
pontos_criticos = solve(derivada_receita, x)

# Avalia a segunda derivada nos pontos críticos para determinar concavidade
segunda_derivada_receita = diff(derivada_receita, x)
concavidade_receita = segunda_derivada_receita.subs(x, pontos_criticos[0])

# Verifica se é um ponto de máximo ou mínimo
if concavidade_receita < 0:
    tipo_ponto = "máximo"
elif concavidade_receita > 0:
    tipo_ponto = "mínimo"
else:
    tipo_ponto = "indefinido"

# Imprime os resultados
print("Quantidade que maximiza a receita:", pontos_criticos[0])
print("Tipo de ponto (mínimo/máximo/indefinido):", tipo_ponto)
```

Pontos críticos são valores de x onde a derivada de uma função é zero ou não está definida.
 Ao analisar a concavidade de uma função nos pontos críticos, podemos determinar se esses pontos correspondem a máximos locais, mínimos locais ou pontos de inflexão.

Vamos considerar uma função f(x) e sua derivada f ′ (x). Os pontos críticos ocorrem quando:  f ′ (x)=0 ou f ′ (x) é indefinida.





*  Se f ′ (x)=0: Um ponto crítico ocorre quando a inclinação da tangente à curva é horizontal. Isso pode indicar um ponto de máximo, mínimo ou ponto de inflexão.
*   Se f′ (x) é indefinida: Em alguns casos, a derivada pode ser indefinida em um ponto crítico, por exemplo, quando a função possui um ponto de descontinuidade.

Para determinar a concavidade nos pontos críticos, usamos a segunda derivada f ′′ (x).





*   Se f" (x)>0: A função é côncava para cima no ponto crítico, indicando um possível ponto de mínimo local.
*   Se f" (x)<0: A função é côncava para baixo no ponto crítico, indicando um possível ponto de máximo local.
*   Se f" (x)=0: O teste da segunda derivada é inconclusivo. Pode ser um ponto de inflexão ou pode indicar que o teste não se aplica nesse ponto.


Portanto, ao examinar a segunda derivada nos pontos críticos, podemos fazer inferências sobre a concavidade da função nessas regiões e determinar se os pontos críticos correspondem a máximos locais, mínimos locais ou pontos de inflexão. Esse é um dos métodos utilizados no teste da segunda derivada para análise de extremos em cálculo diferencial.

```python
from sympy import symbols, diff, solve

# Define a variável simbólica
x = symbols('x')

# Define a função f(x) = x^3 - 6x^2 + 9x + 5
funcao = x**3 - 6*x**2 + 9*x + 5

# Passo 1: Calcula a derivada f'(x)
derivada_primeira = diff(funcao, x)

# Passo 2: Encontra os pontos críticos
pontos_criticos = solve(derivada_primeira, x)

# Passo 3: Calcula a segunda derivada f''(x)
derivada_segunda = diff(derivada_primeira, x)

# Passo 4: Analisa a concavidade nos pontos críticos usando f''(x)
concavidade_pontos_criticos = [derivada_segunda.subs(x, ponto) for ponto in pontos_criticos]

# Imprime os resultados
print("Derivada Primeira (f'(x)):", derivada_primeira)
print("Pontos Críticos:", pontos_criticos)
print("Derivada Segunda (f''(x)):", derivada_segunda)
print("Concavidade nos Pontos Críticos:", concavidade_pontos_criticos)
```

```python
# Define a variável simbólica
x = symbols('x')

# Define a função f(x) = x^3 - 6x^2 + 9x + 5
funcao = x**3 - 6*x**2 + 9*x + 5

# Calcula a segunda derivada f''(x)
derivada_segunda = diff(funcao, x, 2)

# Avalia a segunda derivada nos pontos críticos x = 6 e x = -6
concavidade_ponto_6 = derivada_segunda.subs(x, 6)
concavidade_ponto_neg_6 = derivada_segunda.subs(x, -6)

# Imprime os resultados
print("Segunda Derivada (f''(x)):", derivada_segunda)
print("Concavidade no ponto x = 6:", concavidade_ponto_6)
print("Concavidade no ponto x = -6:", concavidade_ponto_neg_6)
```
