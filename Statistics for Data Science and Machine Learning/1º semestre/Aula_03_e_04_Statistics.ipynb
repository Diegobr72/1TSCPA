# **Normaliza√ß√£o**
##    $n =  \frac{x - x_{min}}{x_{max} - x_{min}} $

```python
#Aplicando a normaliza√ß√£o manualmente:

import numpy as np

X = np.array([10, 20, 30])
xmin = np.min(X)
xmax = np.max(X)

X0 = (X[0] - xmin)/(xmax - xmin)
X1 = (X[1] - xmin)/(xmax - xmin)
X2 = (X[2] - xmin)/(xmax - xmin)
print("Xmax: ", xmax)
print("Xmin: ", xmin)
print("X0: ", X0)
print("X1: ", X1)
print("X2: ", X2)
```

```python
#Automatizando a normaliza√ß√£o:
def normalizacao(x):
    return (x - np.min(x)) / (np.max(x) - np.min(x))


vetorA = np.array(list(range(1,21)))
print(vetorA)

print("\n")#pula linha

vetorA_norm = normalizacao(vetorA)
print(vetorA_norm)
```

# **Padroniza√ß√£o**
##    $p =  \frac{x - \mathbf{ùûµ}}{œÉ} $

```python
def padronizacao(x):
  return (x - np.mean(x))/np.std(x)

vetorB = [10, 20, 30]
vetorB_pad = padronizacao(vetorB)

print(vetorB_pad)
```

## **Sumarizando dados**

```python
import pandas as pd

# Suponhamos que voc√™ tenha um DataFrame de exemplo
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50]
}

df = pd.DataFrame(data)

# Exemplo 1: Aplicar uma √∫nica fun√ß√£o de agrega√ß√£o (m√©dia) a todas as colunas
resultado1 = df.agg('mean')
print("\nExemplo 1:\n", resultado1)

# Exemplo 2: Aplicar v√°rias fun√ß√µes de agrega√ß√£o a uma √∫nica coluna
resultado2 = df['A'].agg(['mean', 'sum', 'min', 'max'])
print("\nExemplo 2:\n",resultado2)

# Exemplo 3: Aplicar fun√ß√µes diferentes a diferentes colunas
resultado3 = df.agg({'A': 'mean', 'B': ['sum', 'min']})
print("\nExemplo 3:\n",resultado3)
```
