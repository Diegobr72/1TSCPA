# **Desvio Padr√£o** \begin{align}
        \mathbf{\sigma}  &= \sqrt{\frac{\sum_{i=1}^n (x_i - ùûµ)^2}{n}}
    \end{align}

```python
x1 = 10
x2 = 9
x3 = 10
x4 = 7.5
x5 = 9.9

media = (x1 + x2 + x3 + x4 + x5)/5

somatorio = (x1 - media)**2 + (x2 - media)**2 + (x3 - media)**2 + (x4 - media)**2 + (x5 - media)**2
desvio_padrao = (somatorio/5)**(1/2)

print("Desvio padr√£o: ", desvio_padrao)
```
