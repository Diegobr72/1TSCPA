## Exercício 1:
Sem utilizar as bibliotecas anteriores, crie uma função que calcule a permutação com e sem repetição, na qual receberá os parâmetros n, k e r, onde
r será um valor booleano que representará se a permutação é com ou sem repetição


```python
def formula_permutacao(n, k, r = False):
  if r is False: #sem repetição, k é apenas um inteiro
    return factorial(n) / factorial(n-k)
  else: #com repetição, k será uma lista contendo [k1, k2, ..., kn]
    divisor = 1
    for i in k:
      divisor = divisor*factorial(i)
    return factorial(n)/divisor
```

```python
formula_permutacao(8, [1,2], True)
```

# **Combinação**

### Exercício 3: Faça uma função que calcule Combinações com e sem repetição

```python
def formula_combinacao(n, k, r=False):
  if r is False:
    return factorial(n)/(factorial(k)*factorial(n-k))
  else:
    return factorial(n+k-1)/(factorial(k)*factorial(n-1))
```

```python
formula_combinacao(8, 4)
```

```python
formula_combinacao(3, 4, True)
```

### Outras aplicações: combinando os looks de um guarda-roupas
Suponha que uma empresa de moda solicitou que você calculasse a quantidade de combinações possíveis de montar looks de um guarda-roupas que tenha 5 chapéus, 10 camisetas, 8 calças e 7 sapatos.

```python
qtd_chapeus = 5
qtd_camisetas = 10
qtd_calcas = 8
qtd_sapatos = 7

qtd_combinacoes = qtd_chapeus*qtd_camisetas*qtd_calcas*qtd_sapatos
qtd_combinacoes
```

### Gerando combinações:

```python
chapeus = ["chapéu "+str(i+1) for i in range(qtd_chapeus)]
camisetas = ["camiseta "+str(i+1) for i in range(qtd_camisetas)]
calcas = ["calça "+str(i+1) for i in range(qtd_calcas)]
sapatos = ["sapato "+str(i+1) for i in range(qtd_sapatos)]
```

```python
guarda_roupa = [chapeus, camisetas, calcas, sapatos]
guarda_roupa
```

```python
len(list(product(*guarda_roupa)))
```

### Organizando tudo em um dataframe

```python
import pandas as pd
combinacoes_roupas = list(product(*guarda_roupa))

df = pd.DataFrame(
    combinacoes_roupas,
    columns=['Chapéu', 'Camiseta', 'Calça', 'Sapato'] )
```

```python
df
```

```python
df = df.reset_index().rename(columns = {'index': 'id_combinacao'})
df
```
